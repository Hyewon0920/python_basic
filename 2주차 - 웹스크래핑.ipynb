```python
def scrape_weather() :
    print("[오늘의 날씨]")
    url ="https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%84%9C%EC%9A%B8+%EB%82%A0%EC%94%A8"
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36","Accept-Language": "ko-KR,ko;q=0.8,en-US;q=0.5,en;q=0.3"}
    res = requests.get(url, headers= headers)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, "lxml")
    summary = soup.find("p", attrs={"class":"summary"}).get_text()
    
    #어제보다 3.5도 높아요 맑음
    #현재온도, 최저 몇 최고 몇
    
    curr_temp = soup.find("div", attrs={"class":"temperature_text"})
    for x in curr_temp :
        print(x.get_text())
        
    #최저온도
    low_temp = soup.find("span", attrs={"class":"lowest"})
    for x in low_temp :
        print(x.get_text())
        
    #최고온도
    high_temp = soup.find("span", attrs={"class":"highest"})
    for x in high_temp :
        print(x.get_text())
        
    #오전/오후 강수확률
    rainfall = soup.find("div", attrs={"class":"cell_weather"})
    for x in rainfall :
        print(x.get_text())
        
    #미세먼지, 초미세먼지
    all_ = soup.find("ul", attrs={"class":"today_chart_list"})
    dust = all_.find_all("li")[1]
    for i in range(2) :
        dust = all_.find_all("li")[i]
        print(dust.get_text())
    
    
```
