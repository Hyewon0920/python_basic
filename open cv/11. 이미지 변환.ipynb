# 14. 이미지 변환 (팽창)

## 이미지를 확장하여 작은 구멍을 채움
흰색 영역의 외곽 픽셀 주변에 흰색을 추가

- dilate

```python
import cv2
import numpy as np

kernnel = np.ones((3,3), dtype = np.uint8)
kernnel

# 이미지 불러오기
img = cv2.imread('dilate.png', cv2.IMREAD_GRAYSCALE) # 흑백으로 불러오기

cv2.imshow('gray', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 확장

import cv2
import numpy as np

kernnel = np.ones((3,3), dtype = np.uint8)
kernnel


img = cv2.imread('dilate.png', cv2.IMREAD_GRAYSCALE) # 흑백으로 불러오기
dilate1 = cv2.dilate(img, kernnel, iterations = 1) # img를 kernnel위에서 iterations : 몇번 반복할지를 정해주는 것 
dilate2 = cv2.dilate(img, kernnel, iterations = 2)
dilate3 = cv2.dilate(img, kernnel, iterations = 3)

cv2.imshow('gray', img)
cv2.imshow('dilate1', dilate1)
cv2.imshow('dilate2', dilate2)
cv2.imshow('dilate3', dilate3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 이미지 침식(팽창 반대)

## 이미지를 깎아서 노이즈 제거
흰색 영역의 외곽 픽셀을 검은색으로 변경

- erode

```python
# 이미지 불러오고 커널 만드는 부분은 똑같음

import cv2
import numpy as np

kernnel = np.ones((3,3), dtype = np.uint8)
kernnel

img = cv2.imread('erode.png', cv2.IMREAD_GRAYSCALE)
erode1 = cv2.erode(img, kernnel, iterations=1) # 1회 반복
erode2 = cv2.erode(img, kernnel, iterations=2) 
erode3 = cv2.erode(img, kernnel, iterations=3) # 함수명만 erode로 바뀌고 나머지는 다 same

cv2.imshow('img', img)
cv2.imshow('erode1', erode1)
cv2.imshow('erode2', erode2)
cv2.imshow('erode3', erode3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 16. 이미지 변환 (열림 & 닫힘)

## 열림 (Opening) : 침식 -> 팽창. 깎아서 노이즈 제거 후 살 찌움
> dilate(erode(image)) -> erode먼저하고 dilate함

```python
# 똑같이 이미지, 커널 불러옴
import cv2
import numpy as np
kernel = np.ones((3, 3), dtype=np.uint8)

img = cv2.imread('erode.png', cv2.IMREAD_GRAYSCALE)

# erode먼저 3회 반복
erode1 = cv2.erode(img, kernnel , iterations = 3)
dilate = cv2.dilate(erode1, kernnel, iterations = 3)

cv2.imshow('img', img)
cv2.imshow('erode1', erode1)
cv2.imshow('dilate', dilate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 결과값



## 닫힘 (Closing) : 팽창 후 침식. 구멍을 메운 후 다시 깎음, 열림과 반대
> erode(dilate(image))

```python
import cv2
import numpy as np
kernel = np.ones((3, 3), dtype=np.uint8)

img = cv2.imread('dilate.png', cv2.IMREAD_GRAYSCALE)

dilate = cv2.dilate(img, kernel, iterations=3)
erode = cv2.erode(dilate, kernel, iterations=3)

cv2.imshow('img', img)
cv2.imshow('dilate', dilate)
cv2.imshow('erode', erode)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
