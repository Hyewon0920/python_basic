# 크기조정

## 이미지 크기조정

1. 고정크기
 - 비율로 설정

```python
import cv2
img = cv2.imread('image.jpg')
dst = cv2.resize(img, (400, 500)) # width, height을 400, 500으로 변경해서 고정해서 출력

cv2.imshow('img', img) # 원본이미지
cv2.imshow('resize', dst) # 바뀐 이미지
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 결과값



비율로 설정
- (400,500)으로 설정해뒀던 걸 none으로 바꿔주기

```python
import cv2
img = cv2.imread('image.jpg')
dst = cv2.resize(img, None, fx=0.5, fy=0.5) # x, y 비율 정의 (0.5 배로 축소-> 절반으로 축소)
# 키울려면 fx=2, fy = 2이런식으로 해주면 됨 원본값은 1,1임

cv2.imshow('img', img)
cv2.imshow('resize', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 결과값



### 보간법

#### 크기를 조정할 때 자연스럽게 보완하는 방법 1-> 10으로 키우면 픽셀도 10으로 커짐 : 깨져보일 수있음
- 이걸 해결하는 방법

1. cv2.INTER_AREA : 크기 줄일 때 사용
1. cv2.INTER_CUBIC : 크기 늘릴 때 사용 (속도 느림, 퀄리티 좋음)
1. cv2.INTER_LINEAR : 크기 늘릴 때 사용 (기본값)

```python
#보간법 적용해 축소

import cv2
img = cv2.imread('image.jpg')
dst = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation =cv2.INTER_AREA ) # x, y 비율 정의 (0.5 배로 축소-> 절반으로 축소)
# 키울려면 fx=2, fy = 2이런식으로 해주면 됨 원본값은 1,1임

cv2.imshow('img', img)
cv2.imshow('resize', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#보간법 적용해 확대

import cv2
img = cv2.imread('image.jpg')
dst = cv2.resize(img, None, fx=2, fy=2, interpolation =cv2.INTER_CUBIC ) # x, y 비율 정의 (0.5 배로 축소-> 절반으로 축소)
# 키울려면 fx=2, fy = 2이런식으로 해주면 됨 원본값은 1,1임

cv2.imshow('img', img)
cv2.imshow('resize', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 동영상 크기 조정

1. 고정크기로 설정할 수 있음

```python
import cv2
cap = cv2.VideoCapture("vd.mp4")

while cap.isOpened() :
    ret, frame = cap.read()
    
    if not ret : 
        break
        
    frame_resized = cv2.resize(frame, (400, 500))        
    cv2.imshow('video', frame_resized)
    #frame_resized = cv2.resize(frame, (400, 500))        
    #cv2.imshow('video', frame_resized)
    #이 두줄은 resize하기 위해 만든거 동영상 resize할려면 이 두줄 추가하면 됨
    # 변경은 frame의 사이즈를 변경하는거임
    if cv2.waitKey(1) == ord("q") :
        break
        
        
cap.release()
cv2.destroyAllWindows()
```
