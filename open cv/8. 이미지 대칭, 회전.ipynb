# 8. 이미지 대칭

1. flip : 이미지 대칭 함수
- flipcode > 0 -> 좌우대칭 so 좌우대칭하고 싶다면 그냥 1 적기
- filpcode == 0 -> 상하대칭 그냥 0 적기
- flipCode < 0 -> 상하좌우 대칭 그냥 -1 적기

## 좌우 대칭

```python
import cv2
img = cv2.imread('image.jpg')
flip_horizontal = cv2.flip(img, 1) # flipCode > 0 : 좌우 대칭 Horizontal

cv2.imshow('img', img)
cv2.imshow('flip_horizontal', flip_horizontal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 상하대칭

```python
import cv2
img = cv2.imread('image.jpg')
flip_horizontal = cv2.flip(img, 0) # flipCode == 0 : 상하 대칭 Horizontal

cv2.imshow('img', img)
cv2.imshow('flip_horizontal', flip_horizontal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 상하좌우 대칭

```python
import cv2
img = cv2.imread('image.jpg')
flip_horizontal = cv2.flip(img, -1) # flipCode < 0 : 상하좌우 대칭 Horizontal

cv2.imshow('img', img)
cv2.imshow('flip_horizontal', flip_horizontal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 이미지 회전

#### rotate라는 함수 사용
- 시계방향으로 90도 회전 : cv2.ROTATE_90_CLOCKWISE
- 시계방향으로 180도 회전 :  cv2.ROTATE_180
- 시계반대방향으로 90도 회전 :  cv2.ROTATE_90_COUNTERCLOCKWISE

## 시계방향 90도

```python
import cv2
img = cv2.imread('image.jpg')

rotate_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) # 시계 방향으로 90도 회전

cv2.imshow('img', img)
cv2.imshow('rotate_90', rotate_90)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 시계방향 180도 회전

```python
import cv2
img = cv2.imread('img.jpg')

rotate_180 = cv2.rotate(img, cv2.ROTATE_180) # 180도 회전

cv2.imshow('img', img)
cv2.imshow('rotate_180', rotate_180)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 시계반대방향 90도 회전

```python
import cv2
img = cv2.imread('img.jpg')

rotate_270 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE) # 시계 반대 방향으로 90도

cv2.imshow('img', img)
cv2.imshow('rotate_270', rotate_270)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
