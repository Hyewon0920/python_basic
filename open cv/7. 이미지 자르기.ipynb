# 7. 이미지 자르기

영역을 잘라서 새로운 윈도우(창)에 표시

```python
# 이미지 가져오기
import cv2
img = cv2.imread("image.jpg")

#가져온 이미지 크기 확인하기
img.shape
```

```python
# 이미지 크기 안에서 잘라야 함
#슬라이싱하는거랑 똑같음
#[]안에 세로, 가로 범위넣어주기
crop = img[100:200, 200:400]
```

```python
# 결과 출력

cv2.imshow("img", img)
cv2.imshow("crop", crop)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 결과값



영역을 잘라서 기존 윈도우에 표시

- crop으로 잘랐던 걸 원래 img에 넣어야 함

```python
import cv2
img = cv2.imread('image.jpg')

crop = img[100:200, 200:400] # 세로 기준 100 : 200 까지, 가로 기준 200 : 400 까지 자름
img[100:200, 400:600] = crop # 원본이미지의 100:200, 400:600까지의 범위 안에 잘라둔 crop을 붙여넣겠단 소리

cv2.imshow('img', img) # 원본 이미지
cv2.waitKey(0)
cv2.destroyAllWindows()
```
