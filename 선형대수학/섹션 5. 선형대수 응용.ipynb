# 주성분 분석





- 기존 데이터를 재정의해서 새로운 데이터 셋을 만든다(기존 데이터로 일차결합해서 단, 새로운 데이터 셋의 분산이 크게끔)
- 차원축소를 한다는 건 지금 그림 상 2차원에 있는 것들을 x축, y축 등 축에 끼워넣는다고 생각하면 됨. 그러면 데이터를 가로지르는 대각선의 축을 찾으면 노란색 파란색 녹색 애들을 잘 끼워넣을 수 있고 그걸 주성분이라고 함

#### 그 축은 가장 큰 고유값에 의해 결정됨

## 방법


1. 데이터를 센터로 끌고 옴 (원래 데이터 평균 구하고 그 데이터에서 평균 빼고 다시 계산)
2. 고유벡터 구하기 위해 공분산 행렬 구하기 (변수에 대한 관계를 나타내는 식 나타내기)
3. 공분산행렬은 대칭 -> orthonormal함
