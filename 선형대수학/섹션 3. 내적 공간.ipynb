# 그람-슈미트 프로세스

## orthonormal set

: 중요! 만약 어떤 s의 벡터의 크기가 전부 1이고 벡터끼리의 내적도 1이라면 그걸 orthonormal set이라고 함

## 그람-슈미트 프로세스

: 유클리디안 공간 V의 basis를 S라고 하면 S가 orthonomal set이면 해당 basis를 orthonormal basis라고 하며 이는 그람-슈미트 프로세스로 구할 수 있다.

## 계산 방법



- S는 orthonormal set이기 이전에 basis이므로 일차결합관계로 표현할 수 있다(c와 v를 이용한)
- 유한직교집합, 서로다른 애들끼리의 내적은 0 이고 자기자신과의 결합은 0이 아님 그렇다면 자기자신과 결합이 있을 딱 1번만 1임(orthonormal이니까)
- 따라서 계수값을 구할 수 있게 됨

## 예시 : 일반 basis를 orthonormal basis로 바꾸기

1. 이는 모든 벡터의 관계를 수직으로 바꾸겠다는 것과 동일하다. 
2. u1 = v1으로 한 점을 고정시키고 그에 맞게 u2를 v2로 옮김(u들을 v로 바꾼다)
3. v1은 u1과 같고 v1에 a를 곱하면 같은 선에 존재할 것임. 그러면 u2와의 관계를 구한 v2가 구해지고 그 두 점의 내적을 구하면 지금 90도가 되었으니 0이고 알파는 스콜라는 밖으로 뺄 수 있고
4. 최종적으로 알파를 구할 수 있게됨
