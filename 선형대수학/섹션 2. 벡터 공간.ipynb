# 1차 독립

### 1차 결합을 했을 때 0이 나오는 경우가 있는데 그때는 a(계수)가 0인 것임 그때의 v들을 일차독립이라고 한다. 

### a1 부터 an까지 하나라도 0이 아닐 때 일차 종속

span : V 벡터에서 추출한 부분집합을 s라고 하고 이 s에 다른 벡터, 스칼라 곱함 -> span s (닫혀있다)




* s는 v1, v2를 가진 벡터 span s는 a1, a2등이 곱해진 s이고 a1, a2는 R에 속한다 = 실수다

예시)



* 지금 a,b,c를 곱한 벡터의 일차결합으로 인해 0벡터가 나옴 그러면 a,b,c도 0이란 얘기 -> ()안의 벡터들은 일차독립이라고 얘기할 수 있음

### 일차독립일 때 다음과 같은 성질을 만족함



* 필요충분 조건 : 1번이면 2번이고 2번이면 1번임
* p -> q면 ~p -> ~q
* 0벡터가 아닌 v들이 일차종속이면 어떠한 벡터도 다른 벡터의 일차결합으로 표현할 수 없다
